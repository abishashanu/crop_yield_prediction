#converting to numeric value and seperating the target value

```python
#Package to work with csv files 
import pandas as pd
```

```python
data=pd.read_csv("/content/yield_df.csv",index_col=0)
```

```python
data.head()
```

```python
area_names=data['Area'].unique()
print(area_names)
#numpy package is for arrays 
import numpy as np
area_number=[]
ar=area_names.tolist() #converting array to list 
for i in data['Area']:
  area_number.append(ar.index(i)) #appending numerical value for area_name
```

```python
#print(area_number)
data.pop('Area')
data['Area_num']=area_number #adding column Area_num
data.head()
```

```python
item_names=data['Item'].unique()
import numpy as np
item_number=[]
item_ar=item_names.tolist()
for i in data['Item']:
  item_number.append(item_ar.index(i))
```

```python
data.pop('Item')
data['Item_num']=item_number
data.head()
```

```python
label=data.pop('hg/ha_yield') #targeted result 
```

```python
data.head()
```

#minmax normalization of data ( all values between 0 to 1 )
# x=(x-minimum)/(maximum-minimum)





```python
for feature_name in data.columns:
        max_value = data[feature_name].max()
        min_value = data[feature_name].min()
        data[feature_name] = (data[feature_name] - min_value) / (max_value - min_value)
```

```python
data.head()
```

#splitting into test and train data (80:20- train:test)

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, label, test_size=0.20, random_state=1) # x-data,y-label 
```
