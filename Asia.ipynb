#converting to numeric value and seperating the target value

```python
#Package to work with csv files 
import pandas as pd
```

```python
data=pd.read_csv("/content/yield_df.csv",index_col=0)
```

```python
data.head()
```

```python
# Importing pandas as pd
from pandas import DataFrame

df = DataFrame(data, columns = ['Area','Item','Year','hg/ha_yield','average_rain_fall_mm_per_year','pesticides_tonnes','avg_temp'])

# Print original data frame
#print("Original data frame:\n")
#print(df)

# Selecting the product of HomeAppliances Type
select_area=df.loc[df['Area'].isin(['Afghanistan','Armenia','Azerbaijan','Bahrain','Bangladesh','Bhutan','Brunei','Cambodia','China','Cyprus','Georgia','India','Indonesia','Iran','Iraq','Israel','Japan','Jordan','Kazakhstan','Kuwait','Kyrgyzstan','Laos','Lebanon','Malaysia','Maldives','Mongolia','Myanmar','Nepal','North Korea','Oman','Pakistan','Palestine','Philippines','Qatar','Russia','Saudi Arabia','Singapore','South Korea','Sri Lanka','Syria','Taiwan','Tajikistan','Thailand','Timor-Leste','Turkey','Turkmenistan','United Arab Emirates','Uzbekistan','Vietnam','Yemen'])]
#print("\n")
# Print selected rows based on the condition
#print("Selecting rows:\n")
print (select_area)
```

```python
area_names=select_area['Area'].unique()
print(area_names)
#numpy package is for arrays 
import numpy as np
area_number=[]
ar=area_names.tolist() #converting array to list 
for i in select_area['Area']:
  area_number.append(ar.index(i)) #appending numerical value for area_name
```

```python
#print(area_number)
select_area.pop('Area')
select_area['Area_num']=area_number #adding column Area_num
select_area.head()
```

```python
item_names=select_area['Item'].unique()
import numpy as np
item_number=[]
item_ar=item_names.tolist()
for i in select_area['Item']:
  item_number.append(item_ar.index(i))
```

```python
select_area.pop('Item')
select_area['Item_num']=item_number
select_area.head()
```

```python
y=select_area.pop('hg/ha_yield') #targeted result 
```

#minmax normalization of data ( all values between 0 to 1 )
x=(x-minimum)/(maximum-minimum)

```python
for feature_name in select_area.columns:
        max_value = select_area[feature_name].max()
        min_value = select_area[feature_name].min()
        select_area[feature_name] = (select_area[feature_name] - min_value) / (max_value - min_value)
```

```python
select_area.head()
```

#splitting into test and train data (80:20- train:test)

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(select_area, y, test_size=0.20, random_state=1) # x-data,y-label 
```
